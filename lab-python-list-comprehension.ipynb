````python
#1 Define the products list
products = ["t-shirt", "mug", "hat", "book", "keychain"]

def initialize_inventory(products):
    """
    Initialize inventory based on the products list.
    
    Parameters:
    - products: List of product names.
    
    Returns:
    - inventory: Dictionary with product names as keys and quantities as values.
    """
    inventory = {product: int(input(f"Enter quantity for {product}: ")) for product in products}
    return inventory

def get_customer_orders():
    """
    Prompt user to enter product names for customer orders.
    
    Returns:
    - customer_orders: Set of ordered product names.
    """
    customer_orders = set()
    while True:
        product = input("Enter the name of the product to order: ").strip().lower()
        if product in products:
            customer_orders.add(product)
        else:
            print(f"{product} is not in the product list.")
        
        another = input("Do you want to add another product? (yes/no): ").strip().lower()
        if another != 'yes':
            break
    return customer_orders

def update_inventory(customer_orders, inventory):
    """
    Update the inventory based on customer orders.
    
    Parameters:
    - customer_orders: Set of ordered product names.
    - inventory: Dictionary with product names as keys and quantities as values.
    """
    for product in customer_orders:
        if inventory.get(product, 0) > 0:
            inventory[product] -= 1

def calculate_order_statistics(customer_orders, products):
    """
    Calculate order statistics.
    
    Parameters:
    - customer_orders: Set of ordered product names.
    - products: List of product names.
    
    Returns:
    - total_products_ordered: Total number of products ordered.
    - percentage_ordered: Percentage of unique products ordered.
    """
    total_products_ordered = len(customer_orders)
    percentage_ordered = (total_products_ordered / len(products)) * 100
    return total_products_ordered, percentage_ordered

def print_order_statistics(order_statistics):
    """
    Print the order statistics.
    
    Parameters:
    - order_statistics: Tuple containing total products ordered and percentage ordered.
    """
    total_products_ordered, percentage_ordered = order_statistics
    print("Order Statistics:")
    print(f"Total Products Ordered: {total_products_ordered}")
    print(f"Percentage of Products Ordered: {percentage_ordered:.2f}%")

def print_updated_inventory(inventory):
    """
    Print the updated inventory.
    
    Parameters:
    - inventory: Dictionary with product names as keys and quantities as values.
    """
    print("Updated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")

# Main program execution
if __name__ == "__main__":
    inventory = initialize_inventory(products)
    customer_orders = get_customer_orders()
    update_inventory(customer_orders, inventory)
    order_statistics = calculate_order_statistics(customer_orders, products)
    print_order_statistics(order_statistics)
    print_updated_inventory(inventory)
````

````python

 
  # 3

def calculate_total_price(customer_orders):
    """
    Calculate the total price of the customer order.
    
    Parameters:
    - customer_orders: Set of ordered product names.
    
    Returns:
    - total_price: Total price of the customer order.
    """
    prices = {product: float(input(f"Enter the price for {product}: ")) for product in customer_orders}
    total_price = sum(prices[product] for product in customer_orders)
    return total_price

def print_order_statistics(order_statistics):
    """
    Print the order statistics.
    
    Parameters:
    - order_statistics: Tuple containing total products ordered and percentage ordered.
    """
    total_products_ordered, percentage_ordered = order_statistics
    print("Order Statistics:")
    print(f"Total Products Ordered: {total_products_ordered}")
    print(f"Percentage of Products Ordered: {percentage_ordered:.2f}%")

def print_updated_inventory(inventory):
    """
    Print the updated inventory using comprehension.
    
    Parameters:
    - inventory: Dictionary with product names as keys and quantities as values.
    """
    print("Updated Inventory:")
    [print(f"{product}: {quantity}") for product, quantity in inventory.items()]

# Main program execution
if __name__ == "__main__":
    inventory = initialize_inventory(products)
    customer_orders = get_customer_orders()
    update_inventory(customer_orders, inventory)
    order_statistics = calculate_order_statistics(customer_orders, products)
    total_price = calculate_total_price(customer_orders)
    print_order_statistics(order_statistics)
    print(f"Total Price of Order: ${total_price:.2f}")
    print_updated_inventory(inventory)
````

````python
def update_inventory(customer_orders, inventory):
    """
    Update the inventory based on customer orders using comprehension.
    Remove the product from the inventory if its quantity becomes zero.
    
    Parameters:
    - customer_orders: Set of ordered product names.
    - inventory: Dictionary with product names as keys and quantities as values.
    """
    # Update the quantities based on customer orders
    inventory.update({product: inventory[product] - 1 for product in customer_orders if product in inventory and inventory[product] > 0})
    
    # Filter out products with quantity of zero
    inventory = {product: quantity for product, quantity in inventory.items() if quantity > 0}
    
    return inventory


def get_customer_orders():
    """
    Prompt user to enter the number of customer orders and gather product names using a loop and comprehension.
    
    Returns:
    - customer_orders: Set of ordered product names.
    """
    num_orders = int(input("Enter the number of products you want to order: "))
    customer_orders = {input(f"Enter the name of product {i + 1}: ").strip().lower() for i in range(num_orders) if input(f"Enter the name of product {i + 1}: ").strip().lower() in products}
    return customer_orders



def print_updated_inventory(inventory):
    """
    Print the updated inventory using comprehension.
    
    Parameters:
    - inventory: Dictionary with product names as keys and quantities as values.
    """
    print("Updated Inventory:")
    [print(f"{product}: {quantity}") for product, quantity in inventory.items()]

# Main program execution
if __name__ == "__main__":
    inventory = initialize_inventory(products)
    customer_orders = get_customer_orders()
    inventory = update_inventory(customer_orders, inventory)
    order_statistics = calculate_order_statistics(customer_orders, products)
    total_price = calculate_total_price(customer_orders)
    print_order_statistics(order_statistics)
    print(f"Total Price of Order: ${total_price:.2f}")
    print_updated_inventory(inventory)
````
